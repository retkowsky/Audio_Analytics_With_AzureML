```python
import numpy as np 
import plotly.graph_objs as go 
import plotly
import IPython
import pandas as pd
import numpy as np
from pyAudioAnalysis import ShortTermFeatures as aF
from pyAudioAnalysis import audioBasicIO as aIO 
from pyAudioAnalysis.audioTrainTest import extract_features_and_train
```

```python
# extract short-term features using a 50msec non-overlapping windows
mt, st = 0.05, 0.05

dirs = ["soundssamples/ok", "soundssamples/ko"]

extract_features_and_train(dirs, mt, mt, st, st, "svm_rbf", "machineanomalymodel")
```

```python
from pyAudioAnalysis import audioTrainTest as aT

cm, thr_prre, pre, rec, thr_roc, fpr, tpr = aT.evaluate_model_for_folders(["soundssamples/ok/", "soundssamples/ko/"], 
                                                                          "machineanomalymodel", 
                                                                          "svm_rbf", 
                                                                          "ok")
```

```python
cm
```

```python
confmatrix = pd.DataFrame(cm, columns = ['c1','c2'])
```

```python
confmatrix
```

```python
TP  = confmatrix.iloc[0,0]
FN  = confmatrix.iloc[0,1]
FP  = confmatrix.iloc[1,0]
TN  = confmatrix.iloc[1,1]
```

```python
accuracy = (TP + TN) / (TP + TN + FN + FP)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1score = 2 * (recall * precision) / (recall + precision)

print("Accuracy =", round(accuracy, 4))
print("Precision =", round(precision, 4))
print("Recall =", round(recall, 4))
print("F1score =", round(f1score, 4))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,7))
plt.clf()
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Greens)

classNames = ['OK','KO']
plt.title('Sound Anomaly Detection')
plt.ylabel('True label')
plt.xlabel('Predicted label')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)

s = [['TP','FN'], ['FP', 'TP']]

for i in range(2):
    for j in range(2):
        plt.text(j,i, str(cm[i][j]))
plt.show()
```

```python
files_to_test = ["test/machine_OK.wav"]

for f in files_to_test:
    print('Analysing ' f'{f}')
    c, p, p_nam = aT.file_classification(f, "machineanomalymodel", "gradientboosting")

    n = np.array(p)
    maxindex = np.argmax(n)
    predictedclass = p_nam[maxindex]
    confidence = max(p)
    print()
    print("\033[1;31;34mPredicted Class:", str.upper(predictedclass))
    print("Confidence =", round(confidence, 5))
```
